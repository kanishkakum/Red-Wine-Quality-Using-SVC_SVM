# Model Training and testing

```python
# separate dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
print(X_train.shape, X_test.shape)
print(Y_train.shape, Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
scalar.fit(X_train)
```

```python
print(scalar.mean_)
```

```python
X_Transform=scalar.transform(X_train)
```

```python
X_Transform
```

```python
Y
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(X_Transform,Y_train)
```

```python
model.score(X_Transform,Y_train)
```

```python
model.predict(X_test)
```

```python
X_test_transform=scalar.transform(X_test)
```

```python
y_pred=model.predict(X_test_transform)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,Y_test)
```

# Hyperparameter Tunning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
#defining the parameter range
param_grid={'C': [0.1,10,100,1000],
           'gamma': [1,0.1,0.01,0.001,0.0001],
           'kernel': ['rbf']}

grid=GridSearchCV(SVC(),param_grid, refit=True, verbose= 3)

```

```python
#fitting the model for grid search
grid.fit(X_Transform,Y_train)
```

```python
#printing the best parameter after the tuning
print(grid.best_params_)
```

```python
#printing how our model looks after the hyperparameter tuning
print(grid.best_estimator_)
```

```python
from sklearn.metrics import classification_report
```

```python
grid_predictions = grid.predict(X_test_transform)

#printing classification report
print(classification_report(Y_test,grid_predictions))
```
